```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
import numpy as np
import sys
np.set_printoptions(threshold=sys.maxsize)

# Define the number of input variables
num_input_variables = 5

# Define the neural network architecture
model = Sequential([
    Dense(20, input_dim=num_input_variables, activation='relu'),
    Dense(16, activation='relu'),
    Dense(8, activation='relu'),
    Dense(8, activation='relu'),
    Dense(5, activation='relu'),
    Dense(1, activation='sigmoid')  # Output layer with 1 neuron (binary classification)
])

# Compile the model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# Print the summary of the model
model.summary()
```

```python
df1 = pd.read_csv('clean.csv')
df2 = pd.read_csv('clean2.csv')
df_temp = pd.concat([df1, df2])
filtered_df_0 = df_temp[df_temp['Label'] != 1]
filtered_df_1 = df_temp[df_temp['Label'] == 1]
df = pd.concat([filtered_df_0[680:], filtered_df_1])
df = df.drop(columns=['Unnamed: 0','Humidity'],axis=1)
df = df.sample(frac = 1)
columns_of_interest = ['Temperature', 'Min Temperature', 'Max Temperature', 'Visibility', 'Wind Speed']

train_data = []
x = 0
for index, row in df.iterrows():
    temp_list = row[columns_of_interest].tolist()
    train_data.append(temp_list)
    x+=1

data = np.array(train_data)
#.reshape((x,5))
```

```python
x = data
Y = np.array(df['Label'].tolist())
```

```python
model.fit(x, Y, epochs=1000, batch_size=128, validation_split=0.2)
```

```python
test_data = []
for index, row in filtered_df_1.iterrows():
    temp_list = row[columns_of_interest].tolist()
    test_data.append(temp_list)
```

```python
predictions = model.predict(np.array(test_data))
```

```python
binary_predictions = (predictions > 0.5).astype(int)
```

```python
x=0
for i in range(len(binary_predictions)):
    if binary_predictions[i] == int(np.array(filtered_df_1['Label'])[i]):
        x+=1
        
print(x/len(binary_predictions))
```
